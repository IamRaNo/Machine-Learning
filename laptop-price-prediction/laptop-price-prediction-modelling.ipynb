```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix
```

```python
X=data.drop('Price',axis=1)
y=data['Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20)
```

```python
X_train.head(1)
```

```python
ohe_encoding=ColumnTransformer(transformers=[
    ('ohe_encoding',OneHotEncoder(sparse_output=False,drop='first'),[0,1,3,9,11,13,15])
],remainder='passthrough')
```

```python
scaling=ColumnTransformer(transformers=[
    ('standard_scaler',StandardScaler(),[3,5,10,12,14,16])
],remainder='passthrough')
```

```python
forest_model=RandomForestRegressor(n_estimators=200,max_features=40)
```

```python
linear_pipe=Pipeline([
    ('encoding',ohe_encoding),
    ('scaling',scaling),
    ('linear_modelling',linear_model)
])
```

```python
forest_pipe.fit(X_train,y_train)
```

```python
forest_prediction=forest_pipe.predict(X_test)
```

```python
r2_score(forest_prediction,y_test)
```
